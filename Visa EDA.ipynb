```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
visa_data = pd.read_csv("./visa.csv")
```

```python
visa_data['Visa_Class'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot('Visa_Class', data = visa_data)
plt.xticks(rotation = 0)
plt.show()
```

```python
explode = [0.4,0.3,0.2, 0.1]
colors = sns.color_palette('bright')
plt.pie(visa_data['Visa_Class'].value_counts(), labels=['H-1B', 'E-3 Australian', 'H-1B1', 'H-1B1'],
        colors = colors, autopct = '%0.0f%%', explode = explode, shadow = 'True',
        startangle = 90, frame = 'true')
plt.show()
```
